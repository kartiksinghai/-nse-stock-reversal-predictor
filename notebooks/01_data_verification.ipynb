```python
START_DATE = "2015-01-01"
END_DATE = "2026-06-22"

test_tickers = [
    "RELIANCE.NS",
    "TCS.NS",
    "INFY.NS",
    "HDFCBANK.NS",
    "ICICIBANK.NS",
    "^NSEI"   # NIFTY 50 index
]

results = []

for ticker in test_tickers:
    try:
        df = yf.download(
            ticker,
            start=START_DATE,
            end=END_DATE,
            interval="1d",
            auto_adjust=False,
            progress=False,
            threads=False
        )

        # Flatten columns if yfinance returns MultiIndex columns
        if isinstance(df.columns, pd.MultiIndex):
            df.columns = [col[0] for col in df.columns]

        results.append({
            "ticker": ticker,
            "status": "OK" if not df.empty else "FAILED_EMPTY",
            "rows": len(df),
            "start_date": df.index.min().date() if not df.empty else None,
            "end_date": df.index.max().date() if not df.empty else None,
            "columns": ", ".join(df.columns.astype(str)) if not df.empty else None
        })

    except Exception as e:
        results.append({
            "ticker": ticker,
            "status": f"ERROR: {type(e).__name__}",
            "rows": 0,
            "start_date": None,
            "end_date": None,
            "columns": str(e)[:150]
        })

verification_report = pd.DataFrame(results)
verification_report
```
